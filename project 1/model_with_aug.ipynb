```python
from fastai.vision.all import *
from fastai.metrics import error_rate
from pathlib import Path
```

```python
# Set the path to your dataset
path = Path('D:/AI Project Guidelines/project 1/data/')
```

```python
# Define the data augmentation transforms
aug_transforms = aug_transforms(do_flip=True, flip_vert=False, max_rotate=45.0, max_zoom=1.1, max_lighting=0.2, max_warp=0.2, p_affine=0.75, p_lighting=0.75)
```

```python
# Create a DataBlock using the ImageBlock and CategoryBlock (for binary classification)
data = ImageDataLoaders.from_folder(path, train='train', valid='valid', item_tfms=Resize(460), batch_tfms=aug_transforms, bs=32)
```
